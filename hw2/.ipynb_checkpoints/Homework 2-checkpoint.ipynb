```python
import requests
import pandas as pd
import numpy as np
from IPython.display import display
from bs4 import BeautifulSoup
import re
```

```python
TOP_UNI_URL = "https://www.topuniversities.com"
```

```python
r = requests.get('https://www.topuniversities.com/sites/default/files/qs-rankings-data/357051_indicators.txt')
json = r.json()['data']
print(json)
```

```python
def getLink(x):
    soup = BeautifulSoup(x, 'html.parser')
    for link in soup.findAll("a"):
        return TOP_UNI_URL + link.get("href")
```

```python
df = pd.DataFrame(json)
df = df[['region','overall_rank','location','uni']]
df['uni'] = df.apply(lambda x: getLink(x['uni']),axis = 1)

display(df)
```

```python
def tist(soup, list_dict_div):
    current_soup = soup
    for dict_div in list_dict_div:
        current_soup = current_soup.find("div",dict_div)
        if(current_soup is None):
            return 0
    return int(current_soup.text.replace(",", ""))
```

```python
def getExtraData(x):
    print(x)
    r_t = requests.get(x)
    soup = BeautifulSoup(r_t.text, 'html.parser')

    total_faculty_int = tist(soup, [{ "class" : "faculty-main wrapper col-md-4" },{ "class" : "total faculty"}, { "class" : "number"}])
    inter_faculty_int = tist(soup, [{ "class" : "faculty-main wrapper col-md-4" },{ "class" : "inter faculty"}, { "class" : "number"}])
    number_students_int = tist(soup, [{"class": "students-main wrapper col-md-4"}, { "class" : "number"}])
    international_students_int = tist(soup, [{"class": "int-students-main wrapper col-md-4"}, { "class" : "number"}])

    return (total_faculty_int,inter_faculty_int,number_students_int,international_students_int)
```

```python
df['total_faculty_int','inter_faculty_int','number_students_int','international_students_int'] = df.apply(lambda x: getExtraData(x['uni']),axis = 1)
```
